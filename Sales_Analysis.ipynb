```python
df_sales['Order Date']= pd.to_datetime(df_sales['Order Date'])
```

```python
df_sales['Order month']=df_sales['Order Date'].dt.month
df_sales['Order year']=df_sales['Order Date'].dt.year
df_sales['Order Day of week']=df_sales['Order Date'].dt.dayofweek
```

```python
Sales_by_month=df_sales.groupby('Order month')['Sales'].sum().reset_index()
Sales_by_month.plot(kind='bar',xlabel='month',ylabel='Sales',title='Monthly Sales Analysis')
plt.show()
```

```python
Sales_by_month=df_sales.groupby('Order Day of week')['Sales'].sum().reset_index()
Sales_by_month.plot(kind='bar',xlabel='week',ylabel='Sales',title='weekly Sales Analysis')
plt.show()
```
